```python
result['arrival_time_minutes'] = result['arrival_time'].map(lambda x: x[3:5])
result['arrival_time_hours'] = result['arrival_time'].map(lambda x: x[:2])
result['departure_time_minutes'] = result['departure_time'].map(lambda x: x[3:5])
result['departure_time_hours'] = result['departure_time'].map(lambda x: x[:2])

result
```

```python
df = pd.read_json('departures.json')
df1 = np.transpose(df).reset_index(drop=True)
```

```python
df1['departures'][0][0]
```

```python
t = []
for i in range(0,len(df1)):
    if df1['departures'][i] == []:
            t.append(i)
df1 = df1.drop(t,axis=0)
```

```python
df1.reset_index(drop=True,inplace=True)
```

```python
type(df1['departures'])
```

```python
est_time =[]
route_id = []
trip_id = []
theo_time = []
headsign = []
df = pd.DataFrame()
for i in range(0,len(df1)):
    est_time.append(df1['departures'][i][0]['estimatedTime'].replace('T',' ').replace('Z', ''))
    route_id.append(df1['departures'][i][0]['routeId'])
    trip_id.append(df1['departures'][i][0]['tripId'])
    theo_time.append(df1['departures'][i][0]['theoreticalTime'].replace('T',' ').replace('Z', ''))
    headsign.append(df1['departures'][i][0]['headsign'])

d = {'estimated_time':est_time,'route_id':route_id,'trip_id':trip_id,'theo_time':theo_time,'headsign':headsign} 
df = pd.DataFrame(data=d)
df
```
